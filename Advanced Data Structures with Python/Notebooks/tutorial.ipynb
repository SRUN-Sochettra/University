Stack

```python
# Creating a stack
def create_stack():
  stack = []
  return stack

# Creating an empty stack
def check_empty(stack):
  return len(stack) == 0

# Adding items into the stack
def push(stack, item):
  stack.append(item)
  print("pushed item: " + item)

# Removing an element from the stack
def pop(stack):
  if (check_empty(stack)):
    return "stack is empty"
  return stack.pop()

stack = create_stack()
push(stack, str(1))
push(stack, str(2))
push(stack, str(3))
push(stack, str(4))
push(stack, str(5))
push(stack, str(6))
print("pop item: " + pop(stack))
print("stack after popping an element: " + str(stack))
```

```python
# Function to reverse a string
def reverse_string(s):
  return s[::-1]

# Example usage for reversing
print("=== REVERSE DATA USING STACK ===")

# Reverse string
text = "National University of Management"
reversed_text = reverse_string(text)
print(f"Original: {text}")
print(f"Reversed: {reversed_text}")
```

```python
# Reverse list
def reverse_list(s):
  return s[::-1]
numbers=[1,2,3,4,5]
reversed_number=reverse_list(numbers)
print(f"Original list: {numbers}")
print(f"Reversed list: {reversed_number}")
```

```python
import queue
import threading
import time

def demonstrate_builtin_queue():
  """ Demonstrate Python's built-in queue implementations """

  print("=== PYTHON BUILT-IN QUEUE ===")

  # 1. FIFO Queue
  print("\n1. FIFO Queue:")
  fifo_queue = queue.Queue(maxsize=3)

  # Add items
  for i in range(3):
    fifo_queue.put(i)
    print(f"Put: {i}")

  # Check if full
  print(f"Queue full: {fifo_queue.full()}")

  # Get items
  while not fifo_queue.empty():
    item = fifo_queue.get()
    print(f"Got: {item}")
    fifo_queue.task_done()

  # 2. LIFO Queue (Stack)
  print("\n2. LIFO Queue (Stack): ")
  lifo_queue = queue.LifoQueue()

  for i in range(3):
    lifo_queue.put(i)
    print(f"Pushed: {i}")

  while not lifo_queue.empty():
    item = lifo_queue.get()
    print(f"Popped: {item}")
```

```python
# 3. Priority Queue
import queue

print("\n3. Priority Queue:")
priority_queue = queue.PriorityQueue()

items = [(3, 'Low'), (1, 'High'), (2, 'Medium')]
for priority, value in items:
  priority_queue.put((priority, value))
  print(f"Added: {value} (priority: {priority})")

while not priority_queue.empty():
  priority, value = priority_queue.get()
  print(f"Processed: {value} (priority: {priority})")
```
